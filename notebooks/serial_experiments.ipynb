```python
import numpy as np

from cytodatagen.markers import NamedMarkerDistribution
from cytodatagen.dists import MultivariateNormal
from cytodatagen.subjects import Subject
from cytodatagen.populations import DistributionPopulation
```

```python
markers = [f"cd_{i}" for i in range(3)]

dist = MultivariateNormal(np.array([1,2,3]), cov=np.eye(3))
mdist = NamedMarkerDistribution(markers, dist)
pop = DistributionPopulation("super_cells", mdist)
subject = Subject("lain", alpha=[10], populations=[pop])

subject.sample(100).obs.head()
```

```python
subject_data = subject.to_dict()
subject_data
```

```python
Subject.from_dict(subject_data)
```
